```python
import json
import os
from pycocotools.coco import COCO

def filter_annotations(annotations_path, image_dir):
    # Load COCO annotations
    coco = COCO(annotations_path)
    # List of image filenames in the train directory
    image_files = set(os.listdir(image_dir))

    # Filter annotations to only include images in `image_files`
    image_ids = [img_id for img_id, img_info in coco.imgs.items() if img_info['file_name'] in image_files]

    # Subset annotations
    filtered_annotations = {'images': [], 'annotations': []}
    for img_id in image_ids:
        filtered_annotations['images'].append(coco.imgs[img_id])
        filtered_annotations['annotations'].extend(coco.imgToAnns[img_id])

    return filtered_annotations

# Usage
annotations_path = '/content/drive/MyDrive/Dataset/annotations/person_keypoints_val2017.json'
image_dir = '/content/drive/MyDrive/Dataset/val'
filtered_annotations = filter_annotations(annotations_path, image_dir)
```

```python
# Save filtered annotations to a new JSON file
def save_filtered_annotations(filtered_annotations, output_path):
    with open(output_path, 'w') as f:
        json.dump(filtered_annotations, f)

# Usage
output_path = '/content/drive/MyDrive/Dataset/filtered_annotations_val.json'
save_filtered_annotations(filtered_annotations, output_path)
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.utils import Sequence
from pycocotools.coco import COCO
import cv2

class PoseDataGenerator(Sequence):
    def __init__(self, annotations_path, image_dir, batch_size=32, input_size=(256, 256), shuffle=True):
        self.coco = COCO(annotations_path)
        self.image_dir = image_dir
        self.batch_size = batch_size
        self.input_size = input_size
        self.shuffle = shuffle
        self.image_ids = list(self.coco.imgs.keys())
        self.on_epoch_end()

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.image_ids)

    def __len__(self):
        return int(np.floor(len(self.image_ids) / self.batch_size))

    def __getitem__(self, index):
        batch_ids = self.image_ids[index * self.batch_size:(index + 1) * self.batch_size]
        images, keypoints = [], []
        for img_id in batch_ids:
            image, keypoint = self.load_image_and_keypoints(img_id)
            images.append(image)
            keypoints.append(keypoint)
        return np.array(images), np.array(keypoints)

    def load_image_and_keypoints(self, img_id):
      img_info = self.coco.loadImgs(img_id)[0]
      image_path = os.path.join(self.image_dir, img_info['file_name'])

      # Load and preprocess the image
      image = cv2.imread(image_path)
      image = cv2.resize(image, self.input_size)
      image = image / 255.0  # Normalize to [0,1]

      # Load and preprocess keypoints
      anns = self.coco.loadAnns(self.coco.getAnnIds(imgIds=img_id))
      keypoints = np.zeros((17, 3), dtype=float)  # COCO has 17 keypoints, set dtype to float
      if anns:
          keypoints_data = anns[0]['keypoints']  # Assume one person per image
          keypoints = np.array(keypoints_data, dtype=float).reshape(17, 3)  # Explicitly set dtype to float

      # Normalize keypoint coordinates
      keypoints[:, :2] /= [img_info['width'], img_info['height']]
      keypoints = keypoints.flatten()  # Flatten to (51,)

      return image, keypoints


# Usage
train_generator = PoseDataGenerator('/content/drive/MyDrive/Dataset/filtered_annotations_train.json', '/content/drive/MyDrive/Dataset/train')
val_generator = PoseDataGenerator('/content/drive/MyDrive/Dataset/filtered_annotations_val.json', '/content/drive/MyDrive/Dataset/val')
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input, Reshape
from tensorflow.keras.models import Model

def build_pose_estimation_model(input_shape=(256, 256, 3), num_keypoints=17):
    inputs = Input(shape=input_shape)

    # Convolutional layers
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    x = MaxPooling2D((2, 2))(x)

    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)

    x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)

    x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)

    x = Flatten()(x)
    x = Dense(512, activation='relu')(x)

    # Output layer for 17 keypoints, each with (x, y, v) format (flattened to 51 values)
    outputs = Dense(num_keypoints * 3, activation='linear')(x)
    model = Model(inputs, outputs)

    return model

# Instantiate and compile the model
model = build_pose_estimation_model()
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
# Train the model with the defined train and validation generators
model.fit(train_generator, validation_data=val_generator, epochs=10)

# Save the trained model
model.save('/content/drive/MyDrive/Dataset/pose_estimation_model.h5')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def preprocess_image(image_path, input_size=(256, 256)):
    # Load the image
    image = cv2.imread(image_path)
    original_shape = image.shape[:2]

    # Resize and normalize
    image_resized = cv2.resize(image, input_size)
    image_normalized = image_resized / 255.0  # Normalize to [0,1]

    # Expand dimensions to match model input shape (1, height, width, channels)
    image_input = np.expand_dims(image_normalized, axis=0)

    return image_input, original_shape

# Example image path
test_image_path = '/content/drive/MyDrive/Dataset/val/000000000139.jpg'
image_input, original_shape = preprocess_image(test_image_path)
```

```python
# Predict keypoints
predicted_keypoints = model.predict(image_input)
predicted_keypoints = predicted_keypoints.reshape(17, 3)  # Reshape to (17, 3) for 17 keypoints with (x, y, v)
```

```python
def visualize_keypoints(image_path, keypoints, original_shape):
    # Load the original image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Scale keypoints back to original image size
    height, width = original_shape
    keypoints[:, 0] *= width
    keypoints[:, 1] *= height

    # Plot the keypoints
    plt.imshow(image)
    for (x, y, v) in keypoints:
        if v > 0:  # Only plot visible keypoints
            plt.plot(x, y, 'ro')  # Red dot for each keypoint
    plt.axis('off')
    plt.show()

# Visualize the predicted keypoints
visualize_keypoints(test_image_path, predicted_keypoints, original_shape)
```

```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.losses import MeanSquaredError

# Specify custom objects to handle the loss function
model = tf.keras.models.load_model(
    '/content/drive/MyDrive/Dataset/pose_estimation_model.h5',
    custom_objects={'mse': MeanSquaredError()}
)

def preprocess_frame(frame, input_size=(256, 256)):
    # Resize and normalize the frame
    frame_resized = cv2.resize(frame, input_size)
    frame_normalized = frame_resized / 255.0
    return np.expand_dims(frame_normalized, axis=0), frame.shape[:2]  # Add batch dimension and return original size

def postprocess_keypoints(keypoints, original_shape):
    height, width = original_shape
    keypoints = keypoints.reshape(17, 3)
    keypoints[:, 0] *= width  # Scale x back to original width
    keypoints[:, 1] *= height  # Scale y back to original height
    return keypoints
```

```python
def real_time_pose_estimation(model, input_size=(256, 256)):
    # Open the default webcam
    cap = cv2.VideoCapture(0)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Preprocess the frame
        frame_input, original_shape = preprocess_frame(frame, input_size)

        # Predict keypoints
        keypoints = model.predict(frame_input)
        keypoints = postprocess_keypoints(keypoints, original_shape)

        # Draw keypoints on the frame
        for (x, y, v) in keypoints:
            if v > 0:  # Check visibility flag
                cv2.circle(frame, (int(x), int(y)), 5, (0, 255, 0), -1)  # Draw green dots for visible keypoints

        # Display the frame with keypoints
        cv2.imshow('Pose Estimation', frame)

        # Exit if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the webcam and close windows
    cap.release()
    cv2.destroyAllWindows()

# Run real-time pose estimation
real_time_pose_estimation(model)
```
